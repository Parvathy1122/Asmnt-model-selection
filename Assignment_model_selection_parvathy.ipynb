```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/titanic_dataset .csv")     #Loading daatset
```

**EDA**

```python
df.describe()
```

```python
df.drop(["PassengerId", "Name","SibSp","Parch","Ticket","Fare","Cabin","Embarked"], axis="columns",inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python

median_age=df['Age'].median()    #Handling missing values
df['Age']= df['Age'].fillna(median_age)
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df)                        #Encoding
```

```python
df.head()
```

```python
x=df.drop('Survived',axis=1)    #diving feature variables and target
y=df['Survived']
print(x)
print(y)
```

**Hold_out technique**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.3)
```

**KNN**

```python

from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=5)

classifier.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)             #model prediction
```

```python
score_KNN = classifier.score(x_test,y_test)
score_KNN
```

```python
from sklearn.metrics import accuracy_score


accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```

**K-FOLD on KNN Model**


```python
from sklearn.model_selection import KFold
Kfold_validator = KFold(10)
for train_index,test_index in Kfold_validator.split(x,y):
    print('training Index:',train_index)
    print('testng Index:',test_index)
```

```python
from sklearn.model_selection import cross_val_score
cv_result= cross_val_score(classifier,x,y,cv=Kfold_validator)
cv_result
```

```python
import numpy as np
np.mean(cv_result)
```

**Stratified Kfold on KNN Model**

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skfold_validator = StratifiedKFold(n_splits=10)
```

```python
for train_index,test_index in skfold_validator.split(x,y):
    print('training Index:',train_index)
    print('testng Index:',test_index)
```

```python
from sklearn.model_selection import cross_val_score
cv_result= cross_val_score(classifier,x,y,cv=skfold_validator)
cv_result
```

```python
import numpy as np
np.mean(cv_result)
```
